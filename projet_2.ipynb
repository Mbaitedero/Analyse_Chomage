```python
# Charger les données de chômage à partir d'un fichier CSV
file_path = 'Unemployment in India (1).csv'
data = pd.read_csv(file_path)

# Afficher les premières lignes du DataFrame pour vérifier le chargement des données
data.head()
```

```python
#Les  dimensions du DataFrame
data.shape
```

```python
#Supprimer les espaces avec strip()
data.columns = data.columns.str.strip()
```

```python
#Afficher  les noms des colonnes
for  col  in data.columns:
    print(col)
```

```python
#Renommer  les colonnes
# Estimated Unemployment Rate (% par Unemployment_Rate',
#'Estimated Employed': 'Employed',
#'Estimated Labour Participation Rate': 'Labour_Participation_Rate',
data.rename(columns={
    'Estimated Unemployment Rate (%)': 'Unemployment_Rate',
    'Estimated Employed': 'Employed',
    'Estimated Labour Participation Rate (%)': 'Labour_Participation_Rate'
}, inplace=True)
data.head()
```

### Voici une explication de chaque colonne du dataset :

### 1. **Region** : La région ou l'état de l'Inde pour lequel les données de chômage sont rapportées.
### 2. **Date** : La date à laquelle les données ont été collectées.
### 3. **Frequency** : La fréquence de collecte des données (par exemple, mensuelle).
### 4. **Estimated Unemployment Rate (%)** : Le taux de chômage estimé en pourcentage pour la région et la période données.
### 5. **Estimated Employed** : Le nombre estimé de personnes employées dans la région pour la période donnée.
### 6. **Estimated Labour Participation Rate (%)** : Le taux de participation estimé de la main-d'œuvre en pourcentage pour la région et la période données.
### 7. **Area** : La zone (urbaine ou rurale) pour laquelle les données sont rapportées.

### Ces colonnes fournissent des informations détaillées sur le chômage et l'emploi dans différentes régions de l'Inde à différentes périodes.

```python
# Créer des boxplots pour détecter les outliers dans les colonnes numériques
numeric_columns = data.select_dtypes(include=['float64']).columns

for col in numeric_columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(y=data[col])
    plt.title(f'Boxplot for {col}')
    plt.show()
```

```python
#Ue  fonction  pour  détecter  les  outliers  dans  les  colonnes  numériques using IQR
def fin_outliers_Iqr(data,  column):
    Q1=data[column].quantile(0.25)
    Q3=data[column].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    outliers=data[(data[column]<lower_bound)|(data[column]>upper_bound)]
    return outliers

    
```
